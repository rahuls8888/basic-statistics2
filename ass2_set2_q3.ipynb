```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#define parameters
mu = 3 #mean
sigma = 2 #standard deviation
#generate random samples
np.random.seed(0)
x1 = np.random.normal(mu, sigma, 10000)
x2 = np.random.normal(mu, sigma, 10000)
#calculate 2x1 and x1 =x2
two_x1 = 2*x1
X1_plus_X2=x1+x2
```

```python
#plot histogram for
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.hist(two_x1, bins=50, density=True, alpha=0.5, color='b', label='2X1')
plt.title('Distributions of 2x1')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
```

```python
#Histogram plot for X1+X2
plt.subplot(122)
plt.hist(X1_plus_X2, bins=50, density=True, alpha=0.5, color='r', label='X1 + X2')
plt.title('Distribution of X1 + X2')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
